## Let's extract year, month and day from the BirthDate Column
### Let's rely on google gemini to find the right REGEX

```python
# prompt: regex expression to extract the year, month and day of birth from the column: BirthDate. the solution *MUST* be a REGEX

import re

def extract_date_parts(date_string):
  match = re.match(r"(\d{4})-(\d{2})-(\d{2})", str(date_string))
  if match:
    year = int(match.group(1))
    month = int(match.group(2))
    day = int(match.group(3))
    return year, month, day
  else:
    return None, None, None

HealthCareDataSet['BirthYear'], HealthCareDataSet['BirthMonth'], HealthCareDataSet['BirthDay'] = zip(*HealthCareDataSet['BirthDate'].apply(extract_date_parts))
```
